```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 10 points, starting at 0 and ending at 2 * pi (~6.28)
x = np.linspace (0, 2 * np.pi, 10)
print(x)
```

```python
y = np.sin(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
resolution = 100
frequency = 2
x = np.linspace(0, 2 * np.pi, resolution)
y = np.sin(x * frequency)
plt.plot(x, y)
```

```python
y1 = np.sin(x * 6)
y2 = np.sin(x * 9)
plt.plot(x, y1 + y2)
```
